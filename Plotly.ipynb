## Q1: Combine the two datsets as one dataframe

```python
x = pd.concat([movies, credits],axis=1)
x
```

## Q2: Remove the following columns:
- 'homepage','tagline','id','overview','status','original_title','movie_id'.

```python
x.drop(columns=['homepage','tagline','id','overview','status','original_title','movie_id'],inplace=True)
```

## Q3: Create Genres dataframe that hase two coulumns Genres and Frequency
- Hint: Use Counter to calculate Frequency
- Add new catagory to Genres called Others if the Frequency < 200 consider

```python
g = x[['genres']]
name = [
    "Action",
    'Adventure',
'Fantasy',
'Science Fiction',
'Crime',
'Drama',
'Thriller',
    'Animation',
'Family',
"Documentary",
'Comedy',
    'Romance',
'Horror',
'Mystery',
"TV Movie"
]
gen = {}
for i  in name:
     gen[i]=0

for i in name:
       gen[i] = x['genres'].str.count(i).sum()

gen
```

```python
genres = pd.DataFrame(data= gen,columns=['genres','frequency'])
genres['genres'] = gen.keys()
genres['frequency'] = gen.values()
genres
```

## Q4: Use bar chart draws the relationship between genres of movies and Frequency.

```python
fig = px.bar(genres, x='genres', y='frequency')
fig.show()
```
